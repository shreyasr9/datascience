```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
recent_grads = pd.read_csv('recent-grads.csv')
clean_data = recent_grads.dropna()
print(recent_grads.shape[0])
print(clean_data.shape[0])
```

```python
fig = plt.figure(figsize=(8,8))
fig.subplots_adjust(wspace=0, hspace=0)

ax1 = fig.add_subplot(2,2,1)
ax1.hist(clean_data['ShareWomen'])
ax1.xaxis.set_visible(False)
ax1.set_ylabel('ShareWomen')
ax1.set_ylim(0,30)
ax1.set_yticklabels([0, 5, 10, 15, 20, 25, 30])

ax2 = fig.add_subplot(2,2,2)
ax2.scatter(clean_data['Unemployment_rate'], clean_data['ShareWomen'])
ax2.xaxis.set_visible(False)
ax2.yaxis.set_visible(False)
ax2.set_xlim(0.0,0.2)

ax3 = fig.add_subplot(2,2,3)
ax3.scatter(clean_data['ShareWomen'], clean_data['Unemployment_rate'])
ax3.set_xlabel('ShareWomen')
ax3.set_ylabel('Unemployment_rate')
ax3.set_xlim(0.0,1.0)
ax3.set_ylim(0.0,0.20)
ax3.set_yticklabels([0.00, 0.05, 0.10, 0.15])
ax3.set_xticklabels([0.0, 0.2, 0.4, 0.6, 0.8], rotation=90)

ax4 = fig.add_subplot(2,2,4)
ax4.hist(clean_data['Unemployment_rate'])
ax4.yaxis.set_visible(False)
ax4.set_xlabel('Unemployment_rate')
ax4.set_xlim(0.0,0.20)
ax4.set_xticklabels([0.00, 0.05, 0.10, 0.15, 0.20], rotation=90)

plt.show()
```

```python
clean_data["ShareMen"] = clean_data["Men"] / clean_data["Total"]
arts = clean_data[clean_data['Major_category']=='Arts']
fig2 = plt.figure(figsize=(8,8))
fig2_ax = fig2.add_subplot(1,1,1)
major_names = arts['Major'].tolist()
locs = np.arange(len(major_names))
offset_locs = locs + 0.35
men_prop = arts['ShareMen'].tolist()
women_prop = arts['ShareWomen'].tolist()
bar_1 = fig2_ax.bar(locs, men_prop, 0.35)
bar_2 = fig2_ax.bar(offset_locs, women_prop, 0.35, color='green')
fig2_ax.set_xticks(offset_locs)
fig2_ax.set_xticklabels(major_names, rotation=90)
plt.legend((bar_1, bar_2), ("ShareMen", "ShareWomen"), loc="upper left")
fig2_ax.grid()
plt.show()
```
